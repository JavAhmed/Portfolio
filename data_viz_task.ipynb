[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
cars_data = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_cars = cars_data[cars_data['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=selected_cars)
plt.title('Revs per Mile for Audi, Hyundai, Suzuki, and Toyota')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:** Suzuki


### 2. Histogram

```python
# Histogram of MPG
plt.hist(cars_data['MPG.city'], label='City MPG')
plt.hist(cars_data['MPG.highway'], label='Highway MPG')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Distribution of MPG in the City and on the Highway')
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** Highway

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=cars_data)
plt.title('Relationship between Wheelbase and Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  Turning circles gets larger


### 4. Barplot

```python
# Barplot for mean horsepower
mean_horsepower_by_type = cars_data.groupby('Type')['Horsepower'].mean().reset_index()
plt.bar(mean_horsepower_by_type['Type'], mean_horsepower_by_type['Horsepower'])
plt.title('Mean Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** Generally yes, large cars have the highest horsepower and small cars have the lowest. However, sporty cars might be the same size as small or compact cars, but have higher horsepower. Vans might be the same size as or bigger than large cars, but have lower horsepower.
